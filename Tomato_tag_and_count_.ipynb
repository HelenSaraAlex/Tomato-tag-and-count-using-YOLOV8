```python
import torch
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def predict_and_display(image_path, model):
    try:
        # Load and preprocess the image
        img_pil = Image.open(image_path)
        img_rgb = img_pil.convert('RGB')

        img_size = 640
        img_resized = img_rgb.resize((img_size, img_size))
        img_np = np.array(img_resized)
        img_norm = img_np / 255.0
        img_torch = torch.from_numpy(img_norm.transpose(2, 0, 1)).float().unsqueeze(0)

        # Make predictions
        with torch.no_grad():
            results = model(img_torch)

        # Initialize counters
        ctr = 0  # Red
        ctg = 0  # Green
        cty = 0  # Yellow
        cti = 0  # Infected (all other classes)

        # Visualize the predictions and count classes
        fig, ax = plt.subplots(figsize=(10, 10))
        ax.imshow(img_rgb)

        # Assuming results is a list of detections
        for detection in results[0].boxes.data:
            class_id = int(detection[5])
            confidence = detection[4]
            box = detection[:4].cpu().numpy()

            # Count classes
            if class_id == 5:  # Red
                ctr += 1
                color = 'red'
            elif class_id == 4:  # Green
                ctg += 1
                color = 'green'
            elif class_id == 6:  # Yellow
                cty += 1
                color = 'yellow'
            else:  # Infected
                cti += 1
                color = 'purple'

            # Draw bounding box and label
            xmin, ymin, xmax, ymax = box
            width, height = xmax - xmin, ymax - ymin
            rect = plt.Rectangle((xmin, ymin), width, height, fill=False, edgecolor=color, linewidth=2)
            ax.add_patch(rect)

            label = f"{model.names[class_id]}: {confidence:.2f}"
            ax.annotate(label, (xmin, ymin - 5), color=color, fontsize=12, ha='left')

        ax.axis('off')
        plt.show()
        # Calculate uninfected tomatoes
        ctu = ctr + ctg + cty - cti

        # Print the counts
        print(f"Red tomatoes: {ctr}")
        print(f"Green tomatoes: {ctg}")
        print(f"Yellow tomatoes: {cty}")
        print(f"Infected tomatoes: {cti}")

    except Exception as e:
        print(f"Error processing image: {e}")
        raise  # This will print the full error traceback

# Usage
image_path = '/content/drive/MyDrive/Nexforce/Tomato_project/valid/images/sample1.jpg'
predict_and_display(image_path, model)
```

```python
import cv2
import torch
import numpy as np
from PIL import Image

def process_frame(frame, model):
    # Preprocess the frame
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img_pil = Image.fromarray(img_rgb)

    img_size = 640
    img_resized = img_pil.resize((img_size, img_size))
    img_np = np.array(img_resized)
    img_norm = img_np / 255.0
    img_torch = torch.from_numpy(img_norm.transpose(2, 0, 1)).float().unsqueeze(0)

    # Make predictions
    with torch.no_grad():
        results = model(img_torch)

    # Initialize counters
    ctr = 0  # Red
    ctg = 0  # Green
    cty = 0  # Yellow
    cti = 0  # Infected

    # Process detections
    for detection in results[0].boxes.data:
        class_id = int(detection[5])
        confidence = detection[4]
        box = detection[:4].cpu().numpy()

        # Count classes
        if class_id == 5:  # Red
            ctr += 1
            color = (0, 0, 255)  # BGR
        elif class_id == 4:  # Green
            ctg += 1
            color = (0, 255, 0)
        elif class_id == 6:  # Yellow
            cty += 1
            color = (0, 255, 255)
        else:  # Infected
            cti += 1
            color = (128, 0, 128)

        # Draw bounding box and label
        xmin, ymin, xmax, ymax = map(int, box)
        cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), color, 2)

        label = f"{model.names[class_id]}: {confidence:.2f}"
        cv2.putText(frame, label, (xmin, ymin - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Calculate uninfected tomatoes
    ctu = ctr + ctg + cty - cti

    # Add count information to the frame
    info = f"Red: {ctr}, Green: {ctg}, Yellow: {cty}, Infected: {cti}, Uninfected: {ctu}"
    cv2.putText(frame, info, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

    return frame

def process_video(input_path, output_path, model):
    # Open the video file
    cap = cv2.VideoCapture(input_path)

    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Process the frame
        processed_frame = process_frame(frame, model)

        # Write the frame
        out.write(processed_frame)

    # Release everything
    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Usage
input_video_path = "/content/drive/MyDrive/Nexforce/Tomato_project/video/vid1.mp4"
output_video_path = "/content/drive/MyDrive/Nexforce/Tomato_project/video"
process_video(input_video_path, output_video_path, model)
```
